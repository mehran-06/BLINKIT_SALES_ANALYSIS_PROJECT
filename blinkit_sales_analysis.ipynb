# BUISNESS REQUIREMENTS

# KPI REQUIREMENTS

```python
#Total Sales: The overall revenue generated from all items sold.
total_sales = df['Sales'].sum()

#Average Sales: The average revenue per sale.
avg_sales = df['Sales'].mean()

#number of Items: The total count of different items sold.
no_of_items = df['Sales'].count()

#Average Rating: The average customer rating for items sold. 
avg_rating = df['Rating'].mean()

print(f"total sales are : {total_sales:,.0f} rupees")
print(f"avg sales are : {avg_sales:,.0f} rupees")
print(f"total items are : {no_of_items } ")
print(f"avg rating is : {avg_rating:,.0f} ")
```

# Chart’s Requirements


```python
#Total Sales by Fat Content:
sales_by_fat = df.groupby('Item Fat Content')['Sales'].sum().reset_index()
sales_by_fat
```

```python
plt.title('Total Sales by Fat Content')
plt.pie(sales_by_fat['Sales'] , labels = sales_by_fat['Item Fat Content'] , autopct = '%0.2f%%' , shadow = True , startangle = 90 )
plt.axis('equal')
plt.show()
```

"Low fat" items generated the highest sales (61% approx).

"Regular" items contributed about 35%.

"Low Fat" (spelled differently) contributed the remaining 4%.

```python
#total Sales by Item Type:
total_sales = df.groupby('Item Type')['Sales'].sum()  # series
total_sales = total_sales.reset_index()  #  converted to a data frame
total_sales = total_sales.sort_values( by='Sales',ascending = False)
total_sales
```

```python
plt.figure(figsize = (15,8))
plt.xlabel('items')
plt.ylabel('sum of sales')
plt.title('total Sales by Item Type:' , fontweight='bold')
plt.xticks(rotation = -90)
bars = plt.bar(total_sales['Item Type'] , total_sales['Sales'] , color = 'skyblue' )
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height,
             f'{height:,.0f}', ha ='center', va='bottom', fontsize=10)
plt.tight_layout()
plt.show()
```

```python
# we can do the same with seaborn
plt.figure(figsize = (15,8))
plt.xlabel('items')
plt.ylabel('sum of sales')
plt.title('total Sales by Item Type:' , fontweight='bold')
bars = sns.barplot(data=total_sales, x ='Item Type', y='Sales')
for bar in bars.patches:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height,
             f'{height:,.0f}', ha ='center', va='bottom', fontsize=10)
plt.xticks(rotation = - 90)
plt.tight_layout()
plt.show()
```

The top-selling categories are:

Fruits and Vegetables (₹178,124)

Snack Foods (₹175,434)

Household (₹135,977)

Lowest sales are in Seafood and Breakfast items.

Perishable everyday-use items seem to drive more revenue.

```python
# Fat Content by Outlet for Total Sales:
grouped = df.groupby(['Outlet Location Type' , 'Item Fat Content'])['Sales']  # series
```

```python
# Fat Content by Outlet for Total Sales:
grouped = df.groupby(['Outlet Location Type' ,'Item Fat Content' ])['Sales'].sum().reset_index()  # data frame
grouped
```

```python
plt.figure(figsize=(16, 8))
ax = sns.barplot(data = grouped ,x='Outlet Location Type', y='Sales', hue ='Item Fat Content' , palette='pastel')
plt.title('Total Sales by Fat Content and Outlet')
plt.xlabel('Outlet')
plt.ylabel('Total Sales')
plt.show()
```

tier 3 outlets recorded the highest sales overall.

In all three tiers, low fat items significantly outperform both Low Fat and Regular.

Indicates a strong preference for health-conscious items across locations.

```python
#Sales by Outlet Size:
sales = df.groupby('Outlet Size')['Sales'].sum().sort_index().reset_index()
sales
```

```python
plt.title('Total Sales by outlet size')
plt.pie(sales['Sales'] , labels = sales['Outlet Size'] , autopct = '%0.2f%%' , shadow = True , startangle = 90 )
plt.axis('equal')
plt.show()
```

Medium-sized outlets contribute the most (42%) to total sales.

Small outlets follow closely (37%).

High-sized outlets account for the least (21%).

Possibly reflects more customer traffic in mid-sized outlets which are more balanced in cost and reach.

```python
#Sales by Outlet Location:
sales = df.groupby('Outlet Location Type')['Sales'].sum().sort_index().reset_index()
sales
```

```python
plt.figure(figsize = (10,5))
sales_by_tiers = sns.barplot(data=sales, x ='Outlet Location Type', y='Sales')
plt.title('Total Sales by outlet location')
plt.xlabel('Outlet location')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()
```

Tier 3 locations generate the highest sales (~₹472K).

Tier 2 and Tier 1 follow in descending order.

Shows that Blinkit is doing particularly well in semi-urban and less metropolitan areas.

```python
# Total Sales by Outlet Establishment
sales = df.groupby('Outlet Establishment Year')['Sales'].sum().sort_index().reset_index()
sales
```

```python
plt.plot( sales['Outlet Establishment Year'] , sales['Sales'] )
plt.title('Sales by Outlet Establishment')
plt.xlabel('Outlet Establishment Year')
plt.ylabel('Sales')
plt.show()
```

Steady distribution of sales across outlets regardless of their establishment year.

Outlets from 1998 and 2017 show strong performance.

Recent outlets (2020–2022) also perform well, suggesting good scalability and adoption of newer branches.
